```python
# Push results back to GitHub
import subprocess
from pathlib import Path

print("\n" + "=" * 50)
print("PUSH TO GITHUB")
print("=" * 50)

# Check if model exists before pushing
model_path = Path("backend/models/shuttlecock_yolo.pt")
model_exists = model_path.exists()

if not model_exists:
    print("✗ Model file not found. Cannot push.")
else:
    print("✓ Model file ready to push")
    print("\nGit status:")
    subprocess.run(["git", "status", "--short"], check=False)
    
    # Ask for confirmation
    print("\n" + "-" * 50)
    confirm = input("\n>>> Push trained model to GitHub? (yes/no): ").strip().lower()
    
    if confirm == "yes":
        print("\nConfiguring git...")
        subprocess.run(["git", "config", "user.email", "colab@training.ai"], check=False)
        subprocess.run(["git", "config", "user.name", "Colab Training"], check=False)
        
        print("Adding files...")
        subprocess.run(["git", "add", "backend/models/shuttlecock_yolo.pt"], check=True)
        
        print("Committing...")
        subprocess.run(["git", "commit", "-m", "Trained shuttlecock YOLO model from Colab"], check=True)
        
        print("Pushing to GitHub...")
        try:
            subprocess.run(["git", "push"], check=True)
            print("\n✓ Successfully pushed to GitHub!")
            print("You can now pull the trained model on your local machine:")
            print("  $ git pull")
        except subprocess.CalledProcessError:
            print("\n✗ Push failed. You may need to authenticate with GitHub.")
            print("Add your personal access token to the repository URL and try again.")
    else:
        print("\n✗ Push cancelled. Model remains only on Colab.")
        print("You can download it manually from Colab.")

print("\n" + "=" * 50)
```

```python
# Clone your GitHub repository
import os
import subprocess

# Get GitHub URL from user
github_url = input("Enter your GitHub repository URL: ").strip()

if not github_url:
    print("✗ No URL provided. Exiting.")
else:
    repo_name = github_url.split("/")[-1].replace(".git", "")
    
    print(f"Cloning {github_url}...")
    subprocess.run(["git", "clone", github_url], check=True)
    
    os.chdir(repo_name)
    print(f"✓ Repository cloned and ready!")
    print(f"Working directory: {os.getcwd()}")
```
